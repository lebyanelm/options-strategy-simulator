```python
import mplfinance as mpf
import pandas as pd
import numpy as np


candles = pd.read_csv(f"./data/EURUSD_5.csv")
candles.columns = ["timestamp", "Open", "High", "Low", "Close", "Volume", "trades"]
candles = candles[["timestamp", "Open", "High", "Low", "Close", "Volume"]]
candles["timestamp"] = candles["timestamp"]*1000
candles["timestamp"] = pd.to_datetime(candles["timestamp"], unit="ms")
candles.set_index("timestamp", inplace=True)
df = candles[-60:]
df
```

```python
# Calculate Bollinger Bands
df = df.copy()
df['SMA'] = df['Close'].rolling(window=3).mean()
df['Upper Band'] = df['SMA'] + 2 * df['Close'].rolling(window=3).std()
df['Lower Band'] = df['SMA'] - 2 * df['Close'].rolling(window=3).std()
```

```python
# Initialize empty arrays filled with NaN for plotting trades and the line
trade_points = np.full(len(df), np.nan)
trade_lines = np.full(len(df), np.nan)

# Define trade points
trade_start_idx = 10  # Index of trade start (e.g., 2024-08-02)
trade_end_idx = 39    # Index of trade end (e.g., 2024-08-04)

# Set the trade points at specific indices
trade_points[trade_start_idx] = df['Close'].iloc[trade_start_idx]
trade_points[trade_end_idx] = df['Close'].iloc[trade_end_idx]

# Set the line data between trade start and end
trade_lines[trade_start_idx:trade_end_idx+1] = df['Close'].iloc[trade_start_idx:trade_end_idx+1]

# Create addplots for trade points and dashed lines
color = "green"
line_width = 1.5

# Create addplots for trade points and straight line
ap_points = mpf.make_addplot(trade_points, scatter=True, markersize=100, marker='^', color=color)
ap_line = mpf.make_addplot(trade_lines, color=color, linestyle='--', width=line_width)

# Define a custom style with opacity for the candlesticks
custom_style = mpf.make_mpf_style(base_mpf_style='charles', 
                        marketcolors={'candle': {'up': 'green', 'down': 'red'},
                                    'edge': {'up': 'k', 'down': 'k'},
                                    'wick': {'up': 'k', 'down': 'k'},
                                    'ohlc': {'up': 'k', 'down': 'k'},
                                    'volume': {'up': '#1f77b4', 'down': '#1f77b4'},
                                    'vcedge': {'up': '#1f77b4', 'down': '#1f77b4'},
                                    'vcdopcod': False,
                                    'alpha': 0.5})


single_point = np.full(len(df), np.nan)

price_level_start = df['Close'].iloc[trade_start_idx]  # Use the close price at that index
price_level_end = df['Close'].iloc[trade_end_idx]  # Use the close price at that index

# Set the single point at the specified index
single_point[trade_start_idx] = price_level_start
single_point

# Plot the candlestick chart with trade points and dashed lines
fig, axlist = mpf.plot(df,
         type='candle',
         style=custom_style,
         title=f"{df.index[trade_start_idx]} to {df.index[trade_end_idx]}",
         addplot=[
            mpf.make_addplot(single_point, scatter=True, markersize=50, marker='o', color=color),
            mpf.make_addplot(df['Upper Band'], color='orange', width=line_width),
            mpf.make_addplot(df['Lower Band'], color='orange', width=line_width)],
         volume=False, returnfig=True)

ax = axlist[0]
ax.axhline(y=price_level_start, color=color, linestyle='--', linewidth=line_width, label='Horizontal Line at 104')
ax.axhline(y=price_level_end, color="red", linestyle='--', linewidth=line_width, label='Horizontal Line at 104')
fig.savefig(fname=f"{df.index[trade_start_idx]} to {df.index[trade_end_idx]}.png")
```
